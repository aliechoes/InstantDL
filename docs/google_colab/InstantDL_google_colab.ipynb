Import the deep learning pipeline from InstantDL

```
from instantdl import GetPipeLine
```

Execute the pipeline with the desired settings, which should be added below in the same order as in the confing.json file. By executing the next line InstantDL will run.

```
P = GetPipeLine('Regression','/content/drive/My Drive/InstantDL/docs/examples/data/Regression',None,1,2,{'save_augmented_images': False, 'resample_images': False,'std_normalization': False},'binary_crossentropy',1,None,True,True)
P.run()
```
